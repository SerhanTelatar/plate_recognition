```python
plates = []

results = model.predict("why.jpg")
# Tahmin edilen bilgileri yazdır
for result in results:
   plates.append(result.boxes.xywh)
   print(result.boxes.xywh)

    
```

```python
import cv2

img = cv2.imread("why.jpg")

print(plates[0])

for plate in plates[0]:
    x, y, w, h, = plate
    x1, x2, y1, y2 = int(x-w/2), int(x+w/2), int(y-h/2), int(y+h/2)
    cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```
